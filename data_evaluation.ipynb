# Data Evaluation for 2017 July dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("2017_july.csv")
data = data[["sensor_id", "location", "lat", "lon", "timestamp", "pressure", "temperature", "humidity"]]
bounding_box = (data.lon.min(), data.lon.max(),
                data.lat.min(), data.lat.max())
print(bounding_box)
borders = (23.2, 23.45, 42.6, 42.75)
data.head()
```

Map of all unique sensor locations:

```python
sofia = plt.imread('map.png')
fig, ax = plt.subplots(figsize = (22,18))
ax.scatter(data.lon, data.lat, zorder=1, alpha= 0.2, c='black', s=60)
ax.set_title('Weather Sensors in Sofia')
ax.set_xlim(borders[0],borders[1])
ax.set_ylim(borders[2],borders[3])
ax.imshow(sofia, zorder=0, extent = borders, aspect= 'auto')
plt.savefig("sensors.png")
```
